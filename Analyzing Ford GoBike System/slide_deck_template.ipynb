## Dataset Overview

> This data set includes information about individual rides made
in a bike-sharing system covering the greater San Francisco Bay area, in 2019.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('fordgobike.csv')
```

## Distribution of Duration (sec)

> The length of the trip is mainly based on values of less than 2000 seconds with a peak of about 600 seconds. The number of Trips values first increases at about 600 seconds from around 8000 values to 12,500 values, but then begins to fall below 2000 values in 2000 seconds

```python
binsize = 500
bins = np.arange(0, df['duration_sec'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = df, x = 'duration_sec', bins = bins)
plt.title('Distribution')
plt.xlabel('Duration sec')
plt.ylabel('Trips')
plt.axis([-500, 10000, 0, 90000])
plt.show();
```

## Average by user type

> I will further explore the dataset by Calculate average trip duration of "Subscribers" and "Customers"

```python
colors=['orange', 'green']
axis = df.groupby('user_type')['duration_sec'].mean().plot(kind='barh', color=colors, figsize=(13,7))
axis.set_title('Average by user type', fontsize=20, y=1.015)
axis.set_ylabel('User type', labelpad=16)
axis.set_xlabel('Duration Sec', labelpad=16);
```

##  duration (sec) distribution per user type and gender

> We can note here that females take longer trips "measured in time" than males and others in both cases. For consumers, the disparity is more evident "13 min for men and others vs 15 for women" than for subscribers.

```python
data = df.query('duration_sec < 3600')
n = sb.catplot(data=data, x='user_type', y="duration_sec", hue="member_gender", 
               kind="violin")


n.set_axis_labels("user Type", "duration sec")
n._legend.set_title('gender')
n.fig.suptitle(' duration per user type and gender',
               y=1.03, fontsize=14, fontweight='semibold');
```
